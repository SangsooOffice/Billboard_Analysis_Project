```python
import pandas as pd
```

```python
df=pd.read_excel('./장르_가사.xlsx')
len(df)
```

```python
df1=pd.read_excel('./빌보드.xlsx')
len(df)
```

```python
result = pd.concat([df1, df], axis=1)
```

```python
result
```

```python
#장르가 미정이거나 가사에 없습니다를 포함한 결측치 데이터 삭제
result=result.loc[df['장르']!='미정']
```

```python
cond=result['가사'].str.contains('없습니다')==False
result=result[cond]
```

```python
len(result)
```

```python
#크롤링할때 잘못 지정해서 바꿈
result['년도'] = result['년도'] + 1
```

```python
result
```

```python
result.to_excel('./billboard_merge.xlsx', index = False)
```
